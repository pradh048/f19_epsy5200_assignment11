#### 3b) The plots in 3a shows multivariate relationship for x column variables and y column variables. Interestingly, the plots in the diagonal axes draw frequency histograms for the variable (here x and y variables are the same). 

```python
# Then run a linear regression, print the summary, and write a couple sentences interpreting the results.
ma_sch_2017 = ma_sch_2017.rename({'Average Salary' : 'avg_sal', '% English Language Learner' : 'eng_lang_learn'}, axis = 1)
income_size = smf.ols(formula = 'avg_sal ~ eng_lang_learn', data = ma_sch_2017).fit()
income_size.summary()
```

#### 3 c) The OLS regression results show that the relationship between Average Salary and % English Language Leaner is weak as shown by the R-squared value of 8.5%; which means that 8.5% of the variance in average salary is explained by the percentage of english language learner variable.

```python
# Problem 4 code here (can split into multiple code chunks if you want)
ma_sch_2017 = ma_sch_2017.rename({'Average Class Size' : 'avg_class_size', '% Economically Disadvantaged' : 'econ_disav', 'School Type': 'school_type'}, axis = 1)
income_size_2 = smf.ols(formula = 'avg_sal ~ eng_lang_learn + avg_class_size + econ_disav', data = ma_sch_2017).fit()
income_size_2.summary()
```
